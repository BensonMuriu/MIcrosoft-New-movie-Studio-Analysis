```python
#Merge dataframes to get data
merged = pd.merge(title_basics,title_principals, on="tconst")
merged_titles = pd.merge(merged,title_akas2, on='tconst')
merged_titles
```

```python
#merge the merged dataframes with tmdb_movies
merged_movies = pd.merge(merged_titles,tmdb_movies,on="title")
merged_movies.head()
```

```python
merged_movies_budget = pd.merge(merged_movies,movie_budgets,left_on='title', right_on='movie')
merged_movies_budget
```

```python
#sort values by popularity in the new merged dataframe
merged_movies_budget.sort_values('popularity', ascending=False)
```

```python
#drop duplicates based on the release dates to get the unique movies
merged_movies_budget.drop_duplicates(subset=['tconst'], inplace=True)
```

```python
#sort values based on popularity of the movie
merged_movies_budget.sort_values('popularity', ascending=False)
```

```python
#merge all files with the name basics columns
merged_all = pd.merge(merged_movies_budget,name_basics, on='nconst')
merged_all.sort_values('popularity', ascending = False)
```

```python
#drop the replicated columns in the data frame
merged_all.drop(['movie','title','original_title_x','original_title_y','ordering_y','id_x','release_date_y'], axis = 1, inplace=True)
```
