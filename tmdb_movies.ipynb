```python
#Get the distribution of popularity against number of movies
#Give a distribution of the figure size
plt.figure(figsize=(10,10))
#Give the plots to be used
plt.scatter(tmdb_movies['popularity'] ,tmdb_movies['vote_count'], c ="blue")
#Give the title, and labels of the axis
plt.title('Number of votes vs popularity')
plt.xlabel('Popularity')
plt.ylabel('Number of votes')
 
# To show the plot
plt.show()
```

```python
#Get the distribution of popularity against vote average
#Give a distribution of the figure size
plt.figure(figsize=(10,10))
#Give the plots to be used
plt.scatter(tmdb_movies['popularity'] ,tmdb_movies['vote_average'], c ="blue")
#Give the title, and labels of the axis
plt.title('Average votes vs popularity')
plt.xlabel('Popularity')
plt.ylabel('Vote average')
 
# To show the plot
plt.show()
```
