```python
movie_gross = pd.read_csv('/home/mungai/Downloads/dsc-phase-1-project-master/zippedData/bom.movie_gross.csv')
movie_gross
```

```python
#check for missing values
movie_gross.info()
```

```python
#calculate percentage of missing data in the data
movie_gross.isna().sum()*100/len(movie_gross)
```

```python
#create a non null column of foreign_gross
non_null = movie_gross[movie_gross['foreign_gross'].notna()]
non_null
```

```python
#change the foreign gross column data from object to integer
non_null = movie_gross['foreign_gross'].replace(',', '').astype(float).astype(int)
type(non_null['foreign_gross'])
```

```python
#find the mean, mode and median of foreign gross
mean = int(non_null['foreign_gross']).mean()
mode = int(non_null['foreign_gross']).mode()
median = int(non_null['foreign_gross']).median()

print({'mean':mean, 'mode':mode, 'median':median})
```
