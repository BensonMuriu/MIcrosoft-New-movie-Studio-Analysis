```python
# drop the columns with more than 75% missing data
title_akas = title_akas.drop(['attributes','language',], axis=1)
```

```python
#drop duplicate title_ids
title_akas2= title_akas.drop_duplicates(subset=["title_id"])
title_akas2
```

```python
#check for data missing after dropping duplicates in percentage form
title_akas2.isna().sum()*100/len(title_akas2)
```

```python
#drop missing values above 75% of data
title_akas2 = title_akas2.drop('types', axis = 1)
```
